```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.filterwarnings('ignore')
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
df = pd.read_csv('/content/data.csv')
```

**Scatter plot**

```python
df.plot(kind = 'scatter',x = 'Pulse', y = 'Calories')
plt.legend(title ='calories per duration')
plt.show()
```

HISTPLOT


```python
plt.figure(figsize = (8,8))
sns.histplot(data = df , x = 'Calories')
plt.legend(title = 'Calories data')
plt.show()
```

LINEPLOT

```python
df['Pulse'].plot(kind='line', figsize=(8, 4), title='Pulse')
plt.gca().spines[['top', 'right']].set_visible(False)
```

3D SCATTERPLOT

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['Pulse'], df['Maxpulse'], df['Calories'])

ax.set_xlabel('Pulse')
ax.set_ylabel('Maxpulse')
ax.set_zlabel('Calories')

plt.show()
```

```python
df.plot(kind='scatter', x='Pulse', y='Maxpulse', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.plot(kind='scatter', x='Duration', y='Pulse', s=32, alpha=.8)
plt.show()
```

```python
df['Calories'].plot(kind='hist', bins=20, title='Calories',color = '#002d9c')
plt.show()
```

```python
df['Maxpulse'].plot(kind='hist', bins=20, title='Maxpulse', color = 'hotpink',edgecolor='black')
plt.show()
```

```python
sns.histplot(bins=10, color='lightgreen', edgecolor='red',kde = True,data = df)
plt.xlabel('Pulse')
plt.ylabel('Duration')
plt.title('Histogram with density plot')
plt.show()
```
